```python
import pandas as pd
import os
import re
import numpy as np
import matplotlib.pyplot as plt

folder_path = '/afs/inf.ed.ac.uk/user/s21/s2103701/Desktop/nls-catalogue-published-material_dc'

file_prefix = 'BIBLIOGRAPHIC_11573881650004341_'

file_end = '.xml'

data = pd.DataFrame()

for i in range(1,52):
    curr_file = file_prefix + str(i) + file_end
    file_path = os.path.join(folder_path, curr_file)

    df = pd.read_xml(file_path)

    data = pd.concat([data, df], ignore_index = True)

data
```

```python
def extract_four_digit_year(value):
    # Check for the presence of 'c', 'ca', or 'circa' at the beginning
    prefix_match = re.match(r'^(c|ca|circa)', str(value), flags=re.IGNORECASE)
    
    # Extract all digits from the value
    digits = re.findall(r'\d', str(value))
    
    # If no digits found or less than 4 digits, return None
    if not digits or len(digits) < 4:
        return None, False
    
    # Join the digits and convert to integer
    year = int(''.join(digits))
    
    # If exactly 4 digits 
    if len(str(year)) == 4 and int(str(year)) > 1000 and int(str(year)) <= 2023:
        if prefix_match:
            return year, False
        if '?' in str(value):
            return year, False
        if not prefix_match:
            return year, True
    # If exactly 4 digits, return the year with True
    elif len(str(year)) == 4 and int(str(year)) > 1000 and int(str(year)) <= 2023:
        return year, True
    # If more than 4 digits, return the original value with False
    elif len(str(year)) > 4:
        return value, False
    # Otherwise, return None with False
    else:
        return None, False

```

```python
def compute_language_proportion(data):
    # Count the occurrences of each language
    language_counts = data['language'].value_counts()

    # Calculate the total number of entries
    total_entries = len(data)

    # Compute the proportion of each language
    language_proportions = language_counts / total_entries

    return language_proportions

data_certain['language'].value_counts()
```

```python
import seaborn as sns

def visualize_century_distribution(data):
    
    # Create subplots with 7 rows (for each century)
    fig, axes = plt.subplots(nrows=7, ncols=1, figsize=(10, 14))
    
    for i in range(4):
        
        century_start = 1401 + i * 100
        
        century = 15 + i
    
        century_data = data[(data.year >= century_start) & (data.year < century_start + 100)]['language'].value_counts()
    
    

        # Plotting the stacked bar chart for the current century
        axes[i].bar(
            x=century_data.index,
            height=century_data.values,
            color=sns.color_palette("viridis", len(century_data)),
            edgecolor='black'
        )
        
        
        
        # Adding labels and title for each subplot
        axes[i].set_title(f'{century}th-Century Language Distributions')
        axes[i].set_ylabel('Number of Collections')
        
        # Rotate x-axis labels for better visibility
        axes[i].tick_params(axis='x', rotation=45)
        
        # Set x-axis tick labels for each subplot
        axes[i].set_xticks(range(len(century_data.index)))
        axes[i].set_xticklabels(century_data.index, rotation=45, ha='right')



    # Adjust layout for better spacing
    plt.tight_layout()

    # Show the plot
    plt.show()
    
visualize_century_distribution(data_certain)
```
