```python
def extract_four_digit_year(value):
    # Check for the presence of 'c', 'ca', or 'circa' at the beginning
    prefix_match = re.match(r'^(c|ca|circa)', str(value), flags=re.IGNORECASE)
    
    # Extract all digits from the value
    digits = re.findall(r'\d', str(value))
    
    # If no digits found or less than 4 digits, return None
    if not digits or len(digits) < 4:
        return None, False
    
    # Join the digits and convert to integer
    year = int(''.join(digits))
    
    # If exactly 4 digits 
    if len(str(year)) == 4 and int(str(year)) > 1000 and int(str(year)) <= 2023:
        if prefix_match:
            return year, False
        if '?' in str(value):
            return year, False
        if not prefix_match:
            return year, True
    # If exactly 4 digits, return the year with True
    elif len(str(year)) == 4 and int(str(year)) > 1000 and int(str(year)) <= 2023:
        return year, True
    # If more than 4 digits, return the original value with False
    elif len(str(year)) > 4:
        return value, False
    # Otherwise, return None with False
    else:
        return None, False

```

```python
def compute_language_proportion(data):
    # Count the occurrences of each language
    language_counts = data['language'].value_counts()

    # Calculate the total number of entries
    total_entries = len(data)

    # Compute the proportion of each language
    language_proportions = language_counts / total_entries

    return language_proportions

data_certain['language'].value_counts()
```

```python
def categorize_languages(language_series, threshold_percentage=0.001):
    # Calculate the total number of collections
    total_collections = language_series.sum()

    # Calculate the threshold count
    threshold_count = total_collections * threshold_percentage
    
    # Identify languages below the threshold and sum their counts
    languages_below_threshold = language_series[language_series < threshold_count]
    count_below_threshold = languages_below_threshold.sum()
    
    # Create a new category "other" and "total" combine counts
    other_total_category = pd.Series(data = [count_below_threshold, total_collections], index=['other', 'total'])
    
    # Remove languages below the threshold from the original series
    language_series = language_series.drop(labels=languages_below_threshold.index)
    
    # Concatenate the original series and the "other" category
    categorized_series = pd.concat([language_series, other_total_category])
    
    return categorized_series



for k in centuries.keys():
    centuries[k] = categorize_languages(centuries[k])
    
centuries
```

```python
from matplotlib import animation

# Sample data (replace this with your data)
d = pd.DataFrame({
    'Year': year,
    'Latin Collections': lat_collections_fifteen
})

# Initialize the figure and axis
fig, ax = plt.subplots()

# Function to update the plot for each frame of the animation
def animate(i):
    ax.clear()
    
    # Calculate cumulative sum of Latin collections up to the current year
    cumulative_sum = d['Latin Collections'][:i+1].cumsum()
    
    # Plot the cumulative sum
    ax.bar(d['Year'][:i+1], cumulative_sum, color='blue')
    
    ax.set_title('Cumulative Number of Latin Collections From 1401 to 1500')
    ax.set_xlabel('Year')
    ax.set_ylabel('Cumulative Number of Collections')

# Create the animation
ani = animation.FuncAnimation(fig, animate, frames=len(d), interval=200)
ani.save('latin_cumulative.mp4', writer=animation.FFMpegWriter(fps=5))

plt.show()
```
